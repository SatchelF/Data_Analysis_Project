```python
import pandas as pd

df = pd.read_csv("Data\Data1_Regular_Conventional.csv", header=2)
df = df.iloc[:-1, :2]
new_headers = ['Date', 'Gas Price']  # Specify your new column names here

df.columns = new_headers
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Months'] = df['Date'].dt.month
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Months'], df['Gas Price'])

# Add labels for the x-axis and y-axis
plt.xlabel('Months')
plt.ylabel('Gasoline Price (USD)')

# Show the plot
plt.show()
```

```python
avg_gas_price = []
for i in range(1,13):
    avg_gas_price.append(df[df['Months'] == i]['Gas Price'].mean())

plt.plot(avg_gas_price)

# Add labels for the x-axis and y-axis
plt.xlabel('Months')
plt.ylabel('Gasoline Price (USD)')

# Show the plot
plt.show()
```
