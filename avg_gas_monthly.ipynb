# average gas prices per month

```python
import pandas as pd

df = pd.read_csv("Data/Data1_Regular_Conventional.csv", header=2)
df = df.iloc[:-1, :2]
new_headers = ['Date', 'Gas Price']  # Specify your new column names here

df.columns = new_headers
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Month'], df['Gas Price'])

# Add labels for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Gasoline Price (USD)')

# Show the plot
plt.show()
```

```python
# avg_gas_price = []
# for i in range(1,13):
#     avg_gas_price.append(df[df['Month'] == i]['Gas Price'].mean())

average_gas_price_monthly = df.groupby('Month')['Gas Price'].mean()

plt.figure(figsize=(10, 6))
plt.plot(average_gas_price_monthly.index, average_gas_price_monthly.values)
plt.title('Monthly Gasoline Prices 1990-2024')
# Add labels for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Gasoline Price (USD)')

plt.grid(True)

# Show the plot
plt.show()
```

```python
df_2010_2024 = df[df['Date'].dt.year >= 2010]
average_gas_price_monthly = df_2010_2024.groupby('Month')['Gas Price'].mean()

plt.figure(figsize=(10, 6))
plt.plot(average_gas_price_monthly.index, average_gas_price_monthly.values)
plt.title('Monthly Gasoline Prices 1990-2024')
# Add labels for the x-axis and y-axis
plt.xlabel('Month')
plt.ylabel('Gasoline Price (USD)')

plt.grid(True)

# Show the plot
plt.show()
```
